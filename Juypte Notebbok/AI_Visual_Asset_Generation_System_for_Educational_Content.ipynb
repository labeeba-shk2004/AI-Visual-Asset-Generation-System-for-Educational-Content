```python
import os
import torch
from diffusers import StableDiffusionPipeline

OUTPUT_DIR = "outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

print("Loading Stable Diffusion pipeline... (first run downloads ~4GB model)")
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")
print("Model loaded.")
```

```python
def generate_library(topics, guidance_scale=7.5, steps=30):
    for topic in topics:
        folder_path = os.path.join(OUTPUT_DIR, topic["folder"])
        os.makedirs(folder_path, exist_ok=True)

        print(f"Generating: {topic['filename']} ({topic['folder']})")
        image = pipe(
            topic["prompt"],
            guidance_scale=guidance_scale,
            num_inference_steps=steps
        ).images[0]

        save_path = os.path.join(folder_path, f"{topic['filename']}.png")
        image.save(save_path)
        print(f"  Saved -> {save_path}\n")

    print(f"Done. {len(topics)} visuals generated in '{OUTPUT_DIR}/'")


if __name__ == "__main__":
    generate_library(TOPICS)
```
